```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize']=[12,8]
```

```python
from skimage import data
```

```python
I = data.chelsea()
```

```python
plt.imshow(I);
```

```python
from skimage import transform
```

```python
A = transform.AffineTransform(scale=(1.5,1.5),rotation=30/180*np.pi,translation=(250,0))
```

```python
print(A.params)
```

```python
B = transform.warp(I,A.inverse,order=0)
```

```python
plt.imshow(B)
```

```python
B = transform.warp(I,A.inverse,order=0,output_shape=(800,1000))
```

```python
plt.imshow(B);
```

```python
plt.imshow(B[300:400,0:100])
plt.title('nearest neighbor');
```

```python
C = transform.warp(I,A.inverse,order=1,output_shape=(800,1000))
```

```python
plt.imshow(C[300:400,0:100])
plt.title('bilinear');
```

```python
D = transform.warp(I,A.inverse,order=3,output_shape=(800,1000))
```

```python
plt.imshow(D[300:400,0:100])
plt.title('bicubic');
```

```python
D.dtype
```

```python
E = abs(C-D)
```

```python
plt.imshow(E[300:400,0:100]*10)
plt.title('difference');
```
