```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['image.cmap']='gray'
plt.rcParams['figure.figsize']=[12,8]
```

```python
from skimage import io
```

```python
I = io.imread('watch.jpg')
```

```python
I.shape
```

```python
plt.imshow(I)
```

```python
N = 255 - I
```

```python
plt.imshow(N)
```

```python
L = np.log(I + 1.0)
```

```python
plt.subplot(1,2,1)
plt.imshow(I)
plt.subplot(1,2,2)
plt.imshow(L)
```

```python
P = I**0.5
```

```python
plt.subplot(1,2,1)
plt.imshow(I)
plt.subplot(1,2,2)
plt.imshow(P)
```

```python
r=np.array((0,150,255))
s=np.array((0,250,255))
```

```python
plt.plot(r,s)
```

```python
x = np.arange(256)
```

```python
T = np.interp(x,r,s).astype('uint8')
```

```python
plt.plot(x,T)
```

```python
C = T[I]
```

```python
plt.subplot(1,2,1)
plt.imshow(I)
plt.subplot(1,2,2)
plt.imshow(C)
```

```python
for i in range(8):
    plt.subplot(3,3,i+1)
    B = np.bitwise_and(I,2**i)
    plt.imshow(B)
plt.subplot(3,3,9)
plt.imshow(I);
```

```python
from skimage.exposure import histogram
```

```python
h, b = histogram(I)
```

```python
plt.bar(b,h)
```

```python
T = (255*np.cumsum(h/sum(h))).astype('uint8')
```

```python
plt.plot(b,T)
```

```python
E = T[I]
```

```python
plt.subplot(1,2,1)
plt.imshow(I)
plt.subplot(1,2,2)
plt.imshow(E)
```

```python
he, be = histogram(E)
```

```python
plt.subplot(1,2,1)
plt.bar(b,h)
plt.subplot(1,2,2)
plt.bar(be,he)
```

```python
r = np.array((0,150,150,255))
s = np.array((1,1,2,2))
```

```python
plt.plot(r,s)
```

```python
hs = np.interp(x,r,s)
```

```python
plt.bar(x,hs)
```

```python
G = (255*np.cumsum(hs/sum(hs))).astype('uint8')
```

```python
plt.plot(x,G)
```

```python
iG = np.interp(x,G,x)
```

```python
plt.plot(x,iG)
```

```python
S = iG[E]
```

```python
plt.subplot(1,2,1)
plt.imshow(E)
plt.subplot(1,2,2)
plt.imshow(S)
```

```python
ht, bt = histogram(S)
```

```python
plt.subplot(1,2,1)
plt.bar(be,he)
plt.subplot(1,2,2)
plt.bar(bt,ht)
```
